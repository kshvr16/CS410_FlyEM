```python
import os
import time
from tqdm.notebook import tqdm
from utils.imageUtils import ImageUtils
from image.controller import ImageController
```

```python
data_dir = "/hpcstor6/scratch01/n/n.lankela001/FlyEM/sample_images"
print(os.listdir(data_dir))
```

```python
working_dir = "/hpcstor6/scratch01/n/n.lankela001/FlyEM/testing_directories"
print(os.listdir(working_dir))
```

```python
experiment = "test01"
```

```python
directories = {
    "start": f"{working_dir}/{experiment}",
    "real": f"{working_dir}/{experiment}/real_images",
    "normalized": f"{working_dir}/{experiment}/normalized_images",
    "cropped": f"{working_dir}/{experiment}/cropped_normalized_images",
    "crop_original": f"{working_dir}/{experiment}/crop_originals",
    "smoothened": f"{working_dir}/{experiment}/smoothened_images",
    "segmented_crops": f"{working_dir}/{experiment}/segmented_crops",
    "original_npy": f"{working_dir}/{experiment}/original_npy",
    "cropped_npy": f"{working_dir}/{experiment}/cropped_npy",
    "masks_npy": f"{working_dir}/{experiment}/masks_npy"
}
```

```python
# Creating each dir as new dir
for keys in tqdm(directories):
    ImageUtils.create_new_dir(directories[keys])
```

```python
# copying the files to the real_images
for files in tqdm(sorted(os.listdir(data_dir))):
    if files.endswith(".tif"):
        ImageUtils.src2dest_copy(data_dir, directories["real"], files)
```

```python
# normalizing all the images
for files in tqdm(sorted(os.listdir(data_dir))):
    if files.endswith(".tif"):
        data = ImageController.read(os.path.join(directories["real"], files))
        normalized_data = ImageController.normalize(data)
        ImageUtils.save_image(os.path.join(directories["normalized"], files), normalized_data)
```

```python
# cropping all the normalized images
for files in tqdm(sorted(os.listdir(directories["normalized"]))):
    if files.endswith(".tif"):
        normalized_data = ImageController.read(os.path.join(directories["normalized"], files))
        cropped_data = ImageController.center_crop(normalized_data)
        ImageUtils.save_image(os.path.join(directories["cropped"], files), cropped_data)
```

```python
# smoothenening all the normalized images
for files in tqdm(sorted(os.listdir(directories["cropped"]))):
    if files.endswith(".tif"):
        normalized_data = ImageController.read(os.path.join(directories["cropped"], files))
        cropped_data = ImageController.center_crop(normalized_data)
        ImageUtils.save_image(os.path.join(directories["smoothened"], files), cropped_data)
```

```python
# getting segmented results from smoothened images
for files in tqdm(sorted(os.listdir(directories["smoothened"]))):
    if files.endswith(".tif"):
        smoothened_image = ImageController.read(os.path.join(directories["smoothened"], files))
        start_time = time.time()
        for threshold in tqdm(range(150, 0, -1)):
            threshold_data = ImageController.threshold(smoothened_image, threshold_value=threshold)
            labeled_data, nr_count = ImageController.label(threshold_data)
            selected_labeled_data = ImageController.select_regions(labeled_data, region_size=1500)
            binary_mask_data = ImageController.binary_mask(selected_labeled_data)
            closed_binary_mask_data = ImageController.close_binary_mask(binary_mask_data)
            segmented_image = ImageController.binary_image(closed_binary_mask_data)
            if ImageController.classify_image(binary_mask_data):
                print("Threshold value: ", threshold)
                ImageController.display(segmented_image)
                ImageUtils.save_image(os.path.join(directories["segmented_crops"], files), segmented_image)
                break
        end_time = time.time()
        print("Total Time Taken: ", (end_time - start_time), " seconds")
        print("---------------------------------------------------------")
```
